目的是写个文件查重工具，帮助用户检测文件内容的重复情况。

通过everything库实现文件列表查询
按照文件大小排序
使用三级判断，1级判断采用文件大小，2级判断采用文件probe哈希值，3级判断采用哈希值
2级哈希采用快速哈希算法可以选择输入采样点数，默认为1，即判断前300个字节，最后300个字节，中间300个字节，中间300个是用文件大小除以2的位置开始算起，采样点数可以设置为1-10个，即最多额外采样10个点，每个点300字节
文件大小小于指定值（默认1MB）的文件，直接采用3级哈希判断
几级判断可以选择自动或者手动，自动默认递增到，3级，即1级判断完之后用2级，2级之后用3级
只用文件大小判断特别容易出错，特别是某些按照固定规则写出来的数据文件，里面的标题和数据保留精度都是一致的，导致文件大小一致，但是内容不一样
